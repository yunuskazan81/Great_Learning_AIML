```python
# dropping the id column
df = df.drop('id', axis=1)

# Replacing categorical vars with actual values
df['room_type'] = df['room_type'].replace({'Shared room': 1, 'Private room': 2, 'Entire home/apt': 3})

df['instant_bookable'] = df['instant_bookable'].replace({'t': 1, 'f': 0})

df['cancellation_policy'] = df['cancellation_policy'].replace({'flexible': 1, 'moderate': 2, 'strict': 3})

df['cleaning_fee'] = df['cleaning_fee'].replace({'True': 1, 'False': 0})

df.head(30)
```

For some reason the cleaning_fee column isn't changing the True & False values to 1 & 0. Didn't have any issue with the other columns. Need to ask why this is.

```python
df.dtypes
```

```python
## for some reason the True and False values in the cleaning fee column didn't change
## using an alternative approach to try and get the 1 & 0 result we want.

def changer(x):
    if x =="True":
        result=1
    else:
        result=0
    return result

df["cleaning_fee"] = df.cleaning_fee.apply(changer)
df.head(20)
```

Seems that the changer turned everything to 0's for some reason. Could there be white space or some other reason as to why the "True" values are not being made 1's?
